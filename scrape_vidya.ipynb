```python
import requests
from bs4 import BeautifulSoup

f = open('articles3.csv', 'wb')
f.write('name,link,author,date \n'.encode())

domain = "https://www.analyticsvidhya.com/blog/category/machine-learning/page/"
for z in range(1, 200):  # as we already done coding for the 1st page
    whole_link = domain + str(z) + "/?utm_source=blog_navbar&utm_medium=machine_learning_button"
    response1 = requests.get(whole_link)
    soupdata = BeautifulSoup(response1.content, 'html.parser')
    info1 = soupdata.findAll('div', {'class': 'list-card-content'})
    for i in info1:
        name1 = i.findAll('h4')
        name = name1[0].text
        link1 = i.findAll('a')
        link = link1[0].get('href')
        Author_date = i.findAll('h6')
        author = Author_date[0].text.split(',')[0].strip()
        date1 = Author_date[0].text.strip().split(',', 2)[1:]
        date = "".join(date1)
        print(name, link, author, date)
        f.write(f"{name},{link},{author},{date} \n".encode())  # \n for going to the next line in the CSV file

f.close()
```

```python
import requests

def fetch_data_from_api(api_url):
    try:
        response = requests.get(api_url)

        if response.status_code == 200:
            data = response.json()
            return data
        else:
            print(f"Failed to fetch data. Status code: {response.status_code}")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error occurred during the API request: {e}")
        return None

def display_api_data(data):
    if data and "entries" in data:
        entries = data["entries"]
        for entry in entries:
            print(f"API Name: {entry['API']}")
            print(f"Description: {entry['Description']}")
            print(f"Auth Required: {entry['Auth']}")
            print(f"Category: {entry['Category']}")
            print(f"API Link: {entry['Link']}")
            print("-" * 50)

def main():
    api_url = 'https://api.publicapis.org/entries'
    data = fetch_data_from_api(api_url)

    if data:
        display_api_data(data)

if __name__ == "__main__":
    main()
```
